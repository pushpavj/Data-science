```python
import numpy as np
import pandas as pd
import scipy.optimize
import matplotlib.pyplot as plt
```

```python
#Creating function model  and create data- y=a*exp((-(x-M)**2)/2*sigma**2)
def guas_func(x,a,b,c):
    y=a*np.exp(-(x-b)**2/(2*c**2))
    return y
```

```python
#Generating clean data
x=np.linspace(0,10,100)
y=guas_func(x,1,5,2)
```

```python
plt.plot(x,y)
```

```python
#Adding noise to the data
yn=y+.2*np.random.normal(size=len(x))
```

```python
plt.plot(x,yn)
```

```python
#Executing fit on the noise data
popt,pcov=scipy.optimize.curve_fit(guas_func,x,yn)
```

```python
popt
```

```python
pcov
```

```python
plt.plot(x,guas_func(x,*popt))
```

```python
plt.plot(x,yn,'g')
plt.plot(x,guas_func(x,*popt),'b--')
```
